#### Manual vs Tool-based MSE Calculation  
This notebook is part of my public #30DaysOfLearning journey.

This code demonstrates how to calculate the Mean Squared Error (MSE), a key cost function in Linear Regression, using two approaches: manual calculation using my function and Scikit-learn‚Äôs built-in function. 

The goal is to make abstract theory in machine learning more tangible by building small, practical projects.

## Python Code

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Create sample data

# Features (independent variable) - e.g., study hours
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)

# Target values (dependent variable) - e.g., test scores
y = np.array([2, 3, 4, 4, 5])
```

```python
# Create and train the model
regression = LinearRegression()
regression.fit(X, y)
```

```python
# Make predictions
y_pred = regression.predict(X)
```

```python
# Calculate MSE manually - part 1
mse_manual_1 = (1/y_pred.size) * sum((y_pred - y) ** 2)

# Calculate MSE manually - part 2
mse_manual_2 = np.average((y_pred - y)**2, axis = 0)

# Calculate MSE using scikit-learn
mse_sklearn = mean_squared_error(y, y_pred)
```

## Display results

```python
print("üìä First Manual MSE:", mse_manual_1)
print("üìä Second Manual MSE:", mse_manual_2)
print("ü§ñ Scikit-learn MSE:", mse_sklearn)

# Let's see if they're all the same!
if np.isclose(mse_manual_1, mse_manual_2) and np.isclose(mse_manual_1, mse_sklearn):
    print("üéâ All three methods give the same result! Consistency for the win!")
else:
    print("üßê Hmm... Something's off. The MSE values don't quite match.")
```

## Plot the data and prediction line

```python
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Prediction line')
plt.title('Linear Regression Model')
plt.xlabel('X (Feature)')
plt.ylabel('y (Target)')
plt.legend()
plt.grid()
plt.show()
```
