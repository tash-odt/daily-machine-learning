## Python Code

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function and its derivative
def f(x):
    return x**2

def df(x):
    return 2 * x

# Gradient descent parameters
current_position = 10
step_multiplier = 0.1
precision = 0.001

# Store positions for visualization
positions = [current_position]

# Gradient descent loop
for n in range(500):
    previous_position = current_position
    gradient = df(previous_position)
    current_position = previous_position - step_multiplier * gradient
    positions.append(current_position)

    step_size = abs(current_position - previous_position)
    if step_size < precision:
        break

# Final results
print('Local minimum occurs at x =', current_position)
print('Slope or df(x) value at this point is:', df(current_position))
print('f(x) value or cost at this point is:', f(current_position))
```

## Visualize the code

```python
# Visualization
x = np.linspace(-11, 11, 400)
y = f(x)

# Plot the cost function
plt.plot(x, y, label='f(x) = xÂ²')

# Plot the descent path
list_x = positions
list_y = [f(x) for x in positions]
plt.plot(list_x, list_y, 'ro-', label='Gradient Descent Path')

# Labels and legend
plt.title('Gradient Descent on f(x) = xÂ²')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

## N.B:
In gradient descent, the x values (positions) are calculated step by step during the algorithm. They are not predefined like the x values in the smooth curve.


For visualization, we first define a range of x values (using np.linspace) to show the shape of the cost function. Then, for each of these x values, we calculate the corresponding y values to plot the smooth curve.
