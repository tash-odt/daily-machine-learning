## Python Code

```python
# Import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
%matplotlib inline
```

## Preview data
Initially previewed the CSV file containing the sample data — only for reference, not for use.

The actual data was provided directly as a NumPy array instead.

```python
data = pd.read_csv('age_height.csv')
print(data)

# predicting age based on height is an unrealistic simplifcation of reality --but this is simply for learning sake
```

```python
# Sample data: Height (in inches) vs. Age (in years)
# Defining the height data as a NumPy array (in inches)
# We reshape it to make it a 2D array with one column (required for scikit-learn)
height = np.array([152.40, 157.48, 162.56, 165.10, 172.72, 177.80, 182.88, 187.96, 190.5, 198.12]).reshape(-1, 1)

# Defining the corresponding age data as a NumPy array (in years)
age = np.array([12, 13, 16, 19, 22, 28, 18, 35, 28, 30])
```

### Explanation:
- Height: This is the independent variable (feature) representing the height of individuals in inches. 

The .reshape(-1, 1) is used to convert it into a 2D array (with one feature column), which is the format required by scikit-learn for training.


- Age: This is the dependent variable (target) representing the age of individuals in years. It's a 1D array corresponding to the heights.


In this dataset, the goal is to predict age based on height using linear regression.

```python
# Creating a linear regression model using scikit-learn
regression = LinearRegression()

# Training the model with the data
regression.fit(height, age)

# Making predictions
age_predictions = regression.predict(height)

# Plotting the data and the regression line
plt.scatter(height, age, color='blue', label='Actual data')
plt.plot(height, age_predictions, color='red', label='Regression line')
plt.xlabel('Height (cm)')
plt.ylabel('Age (years)')
plt.title('Linear Regression: Height vs. Age')
plt.legend()
plt.grid()
plt.show()


# Calculating the Residual Sum of Squares (RSS)
residuals = age - age_predictions
rss = np.sum(residuals ** 2)

# Printing the RSS value
print('Residual Sum of Squares (RSS):', rss)

# Calculating Mean Squared Error (MSE) for comparison
mse = mean_squared_error(age, age_predictions)
print('Mean Squared Error (MSE):', mse)

# Calculate the R-squared score
r_squared = r2_score(age, age_predictions)
print('R-squared:', r_squared)

# Interpreting the R² score:
# R² = 1: Perfect fit. The model explains 100% of the variance in the data.
# R² = 0: The model doesn't explain any of the variance.
# 0 < R² < 1: The model explains a proportion of the variance. 
# A higher R² indicates a better fit. 
```

# Key Takeaways
- Scikit-learn’s LinearRegression fits a best-fit line by minimizing the Residual Sum of Squares (RSS) — its built-in cost function.

- RSS measures total error between actual and predicted values. Lower RSS means better predictions.

- MSE (Mean Squared Error) is a related metric that averages the squared errors.

- Plotting helps visualize the model’s fit against the actual data.
